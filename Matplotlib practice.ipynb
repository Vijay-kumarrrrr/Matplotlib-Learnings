```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Basic

```python
x = [0,1,2,3,4,5]
y = [2,4,6,8,10,12]
z = np.arange(0,10,1.8)
```

```python
# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)

plt.figure(figsize = (15,7), dpi = 100)

plt.plot(x,y, 'b*--', label = 'Fx1')
# plotting a line dashed and continuous form
plt.plot(z[:5],z[:5]**2, 'r*--', label = 'Fx2')
plt.plot(z[4:],z[4:]**2, 'r*-')
plt.xlabel('x')
plt.ylabel('y')
plt.title("First Graph", fontdict = {'fontname':'Arial'})
plt.legend()
plt.savefig('graph.png', dpi = 280)
plt.show()
```

# Bar Chart

```python
Attributes = ['X','Y','Z']
Values = [4,12,6]

plt.figure(figsize = (4,6), dpi = 150)

bars = plt.bar(Attributes,Values)

patterns = ['-','*','/']

for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.show()
```

# line graph

```python
gas = pd.read_csv('gas_prices.csv')
gas.head(10)
```

```python
plt.figure(figsize = (18,10))

plt.title('GAS Prices in USD of time')

plt.plot(gas['Year'],gas['USA'],'r*-', label = 'USA')
plt.plot(gas['Year'],gas['Germany'],'g*-', label = 'Germany')
plt.plot(gas.Year,gas.France,'b*-', label = 'France')
plt.plot(gas.Year,gas.Mexico,'c*-', label = 'Mexico')
plt.plot(gas.Year,gas.Italy,'y*-', label = 'Italy')
plt.plot(gas.Year,gas.Japan,'m*-', label = 'Japan')
plt.plot(gas.Year,gas['South Korea'],'y.-', label = 'Sk')
plt.plot(gas.Year,gas.Australia,'k*-', label = 'Australia')
plt.plot(gas.Year,gas.UK,'m.-', label = 'UK')
plt.plot(gas.Year,gas.Canada,'g.-', label = 'Canada')
plt.xlabel('year')
plt.ylabel('USD')
plt.legend()
plt.show()
```

```python
#Another Way to plot many values!
countries = ['Australia', 'USA', 'Canada', 'South Korea','UK','Japan','Germany','Italy','France','Mexico']
for country in gas:
    if country in countries:
        plt.plot(gas.Year, gas[country], marker='.')
```

# Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count(0)[1]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count(0)[2]

plt.figure(figsize = (5,5))

labels = ['Left','Right']
color = ['#acbdea','#bcaade']

plt.pie([left,right], labels = labels, colors = color, autopct = '%.3f')
plt.title('Foot Prefference of FIFA players')
plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.style.use('ggplot')
lesser = fifa[fifa.Jumping < 35].count()[0]
less = fifa[(fifa.Jumping >= 35) & (fifa.Jumping < 50)].count()[0]
med = fifa[(fifa.Jumping >= 50) & (fifa.Jumping < 65)].count()[0]
high = fifa[(fifa.Jumping >= 65) & (fifa.Jumping < 80)].count()[0]
higher = fifa[fifa.Jumping >= 80].count()[0]

Jump = [lesser,less,med,high,higher]
label = ['below 35','35-50','50-65','65-80','greater than 80']

explode = (.4,.2,0,.2,0)

plt.title('Jumps of Professional Soccer Players')
plt.pie(Jump, labels = label, explode = explode, autopct = '%.4f')

plt.show()
```

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```
